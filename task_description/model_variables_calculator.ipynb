```python
calc = lambda k, l, m, n: m * k + l * k + l * (m * k) + l * n + l * n
```

```python
def with_l(k, l, m, n, find):
    result = False
    count = calc(k, l, m, n)
    if count == find:
        message = '{},{},{},{},{}'.format(count,m, k, l, n)
        print(message)
        result = True
    return result
```

```python
def with_n(k, n, m, find):
    result = False
    max_value = 2 * n
    for l in range(2, max_value + 1):
        if with_l(k, l, m, n, find):
            result = True
            break
    return result
```

```python
def with_k(k, m, find):
    result = False
    for n in range(1, 2 * m + 1):
        if with_n(k, n, m, find):
            result = True
            break
    return result
```

```python
def with_m(m, find):
    result = False
    for k in range(2, m):
        if with_k(k, m, find):
            result = True
            break
    return result
```

```python
def search(find):
    max_m = 100
    for m in range(10, max_m + 1):
        if with_m(m, find):
            break
    else:
        raise Exception('Not found for', find)
```

```python
for find in range(100, 20_000 + 1, 100):
    search(find)
```
