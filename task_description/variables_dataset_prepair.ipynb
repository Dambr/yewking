```python
def with_l(k, l, m, n, find):
    result = False
    model_size_calculator = ModelSizeCalculator(k, l, m, n)
    var_count = model_size_calculator.calculate_variables_count()
    con_count = model_size_calculator.calculate_constraints_count()
    if var_count == find:
        message = '{},{},{},{},{},{}'.format(var_count, con_count, m, k, l, n)
        print(message)
        # result = True
    return result
```

```python
def with_n(k, n, m, find):
    result = False
    # max_value = 2 ** n
    for l in range(2, n + 1):
        if with_l(k, l, m, n, find):
            result = True
            break
    return result
```

```python
def with_k(k, m, find):
    result = False
    for n in range(2, 2 * m + 1):
        if with_n(k, n, m, find):
            result = True
            break
    return result
```

```python
def with_m(m, find):
    result = False
    for k in range(2, m):
        if with_k(k, m, find):
            result = True
            break
    return result
```

```python
def search(find):
    max_m = 25
    for m in range(10, max_m + 1):
        if with_m(m, find):
            break
    # else:
    #     raise Exception('Not found for', find)
```

```python
for find in range(100, 20_000 + 1, 100):
    search(find)
```

```python
rows = content.split('\n')
```

```python
rows = rows[1:]
```

```python
rows = rows[:-1]
```

```python
rows = [el.split(',') for el in rows]
```

```python
for i in range(len(rows)):
    cur_row = rows[i]
    for j in range(len(cur_row)):
        rows[i][j] = int(rows[i][j])
```

```python
for row in rows:
    row.append(False)
```

```python
def find_last_with_false(i):
    while i > 0:
        i -= 1
        if rows[i][6] == False:
            return i
    raise Exception('incorrect state, i =', i)
```

```python
for i in range(1, len(rows)):
    last_with_false = find_last_with_false(i)
    if rows[i][1] < rows[last_with_false][1]:
        rows[i][6] = True
```
