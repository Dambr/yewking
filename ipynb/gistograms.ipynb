```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
steps = range(0, 16)
```

```python
hashset = [42, 52, 60, 41, 57, 77, 63, 65, 7830, 7402, 8364, 8163, 6932, 6807, 8191, 8496]
linkedhashset = [37, 52, 40, 29, 88, 48, 52, 71, 4061, 2894, 3065, 3083, 2755, 2892, 3017, 3415]
treeset = [65, 71, 70, 52, 51, 58, 78, 119, 8083, 6264, 7509, 7562, 6640, 6911, 7982, 7705]

ax = plt.subplot(111)

for i in range(len(steps)):
    ax.bar(1 + steps[i] - 0.2, hashset[i], width=0.2, color='#Add8e6', align='center')
    ax.bar(1 + steps[i], linkedhashset[i], width=0.2, color='#90ee90', align='center')
    ax.bar(1 + steps[i] + 0.2, treeset[i], width=0.2, color='#ff7f7f', align='center')

colors = {'TreeSet':'#ff7f7f', 'LinkedHashSet':'#90ee90', 'HashSet':'#Add8e6'}         
labels = list(colors.keys())
handles = [plt.Rectangle((0,0),1,1, color=colors[label]) for label in labels]
plt.legend(handles, labels)
plt.xlabel("Номер эксперимента")
plt.ylabel("Среднее время, мс")

plt.show()
```
